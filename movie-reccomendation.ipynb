```python
m = df['vote_count'].quantile(0.9)

average = df['vote_average'].mean()

rating = []

for i, j in zip(df['vote_average'], df['vote_count']):
    k = (j * i) + (m * average)
    rating.append(k/(j + m))
    
df['W_A'] = rating

df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
df['overview'] = df['overview'].fillna('')
tfidf_matrix = tfidf.fit_transform(df['overview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def indexofthe(watched_movie, feature):
    k = 0
    for i in feature:
        if i == watched_movie:
            return k
        k+=1
    return None
```

```python
def reccomendation(watched_movie, cosine_sim = cosine_sim):
    
    
    indexed = indexofthe(watched_movie, df['original_title'])
    cosine_sim[indexed][indexed] = -1
    k = cosine_sim[indexed].copy()
    k.sort()
    other_movies_score = k
    other_movies_score = other_movies_score[-26:-1][::-1]
    indexes = []

    for i in other_movies_score:
        index_of_reccomended_movie = indexofthe(i, cosine_sim[indexed])
        indexes.append(index_of_reccomended_movie)

    return df['original_title'].iloc[indexes]
```

```python
import pandas as pd, numpy as np
from ast import literal_eval

movies = pd.read_csv('/kaggle/input/datasets/organizations/tmdb/tmdb-movie-metadata/tmdb_5000_movies.csv')
credits = pd.read_csv('/kaggle/input/datasets/organizations/tmdb/tmdb-movie-metadata/tmdb_5000_credits.csv')
credits.columns = ['id', 'title', 'cast', 'crew']
smd = movies.merge(credits, on='id')

for f in ['cast', 'crew', 'keywords', 'genres']:
    smd[f] = smd[f].apply(literal_eval)
```

```python
links = pd.read_csv('/kaggle/input/datasets/rounakbanik/the-movies-dataset/links_small.csv')
links = links[links['tmdbId'].notnull()]['tmdbId'].astype('int')

smd = smd[smd['id'].isin(links)]
```

```python
id_map = pd.read_csv('/kaggle/input/datasets/rounakbanik/the-movies-dataset/links.csv')[['movieId', 'tmdbId']]
id_map.columns = ['movieId', 'id']
id_map = id_map.merge(smd[['title_x', 'id']], on='id').set_index('title_x')

indices_map = id_map.set_index('id')
```

```python
from surprise import Reader, Dataset, SVD

reader = Reader()
ratings = pd.read_csv('/kaggle/input/datasets/rounakbanik/the-movies-dataset/ratings_small.csv') 
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

svd = SVD()
trainset = data.build_full_trainset()
svd.fit(trainset)
```

```python
smd = smd.reset_index(drop=True)
smd['overview'] = smd['overview'].fillna('')

tfidf_smd = TfidfVectorizer(stop_words='english')
tfidf_matrix_smd = tfidf_smd.fit_transform(smd['overview'])
cosine_sim = linear_kernel(tfidf_matrix_smd, tfidf_matrix_smd)

indices = pd.Series(smd.index, index=smd['title_x']).drop_duplicates()
```

```python
def hybrid(userId, title):
    idx = indices[title]
    if hasattr(idx, 'iloc'):        
        idx = idx.iloc[0]

    sim_scores = list(enumerate(cosine_sim[int(idx)]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:26]
    movie_indices = [i[0] for i in sim_scores]

    result = smd.iloc[movie_indices][['title_x', 'vote_count', 'vote_average', 'id']].copy()
    result['est'] = result['id'].apply(
        lambda x: svd.predict(userId, indices_map.loc[x]['movieId']).est)
    result = result.sort_values('est', ascending=False)
    return result.head(10)
```
